# Создание датасета

```python
1. Определите формат 3D-моделей мебели, который вы хотите создать, например, .obj.

2. Создайте скрипт, который будет генерировать случайные модели мебели в выбранном формате. Для этого вы можете использовать Open3D.

3. Создайте скрипт для преобразования моделей в изображения. Это может включать в себя настройку параметров рендеринга, например, размер изображения, положение источника света и т.д.

4. Создайте скрипт для генерации соответствующих аудиофайлов. Вы можете использовать библиотеку Gtts для генерации речи на основе текста.

5. Объедините изображения и соответствующие аудиофайлы в единый датасет, используя, например, формат .csv.
```

В конце нам нужно разделиить датасет на обучающую и тестувую выборку.

```python
import open3d as o3d
import random

# Количество случайных моделей, которые нужно сгенерировать
num_models = 100

# Список названий моделей мебели
furniture_names = ["стул", "диван", "стол", "кровать", "дверь"]

# Параметры для создания случайных моделей
# Можно изменять эти параметры, чтобы получать разные модели
chair_params = {'height': 0.8, 'width': 0.4, 'depth': 0.4}
sofa_params = {'height': 0.9, 'width': 2.0, 'depth': 0.8}
table_params = {'height': 0.7, 'width': 1.0, 'depth': 0.6}
bed_params = {'height': 1.2, 'width': 2.0, 'depth': 1.5}
door_params = {'height': 2.2, 'width': 0.9, 'depth': 0.05}

# Список параметров для каждой модели мебели
furniture_params = [chair_params, sofa_params, table_params, bed_params, door_params]

# Генерация случайных моделей мебели
for i in range(num_models):
    # Выбор случайной модели мебели
    furniture_idx = random.randint(0, len(furniture_names)-1)
    furniture_name = furniture_names[furniture_idx]
    furniture_param = furniture_params[furniture_idx]
    
    # Создание случайной модели мебели
    if furniture_name == 'стул':
        mesh = o3d.geometry.TriangleMesh.create_box(width=furniture_param['width'], 
                                                    height=furniture_param['height'], 
                                                    depth=furniture_param['depth'])
    elif furniture_name == 'диван':
        mesh = o3d.geometry.TriangleMesh.create_box(width=furniture_param['width'], 
                                                    height=furniture_param['height'], 
                                                    depth=furniture_param['depth'])
    elif furniture_name == 'стол':
        mesh = o3d.geometry.TriangleMesh.create_box(width=furniture_param['width'], 
                                                    height=furniture_param['height'], 
                                                    depth=furniture_param['depth'])
    elif furniture_name == 'кровать':
        mesh = o3d.geometry.TriangleMesh.create_box(width=furniture_param['width'], 
                                                    height=furniture_param['height'], 
                                                    depth=furniture_param['depth'])
    elif furniture_name == 'дверь':
        mesh = o3d.geometry.TriangleMesh.create_box(width=furniture_param['width'], 
                                                    height=furniture_param['height'], 
                                                    depth=furniture_param['depth'])
    
    # Сохранение модели мебели в формате .obj
    file_name = f'{furniture_name}_{i}.obj'
    o3d.io.write_triangle_mesh(file_name, mesh)
```

здесь нужно посмотреть что сделать чтобы преобразовать модели в изображения '.png'.

```python
import open3d as o3d
import numpy as np

# Размеры изображений
image_width = 480
image_height = 320

# Положение источника света
light_source = o3d.geometry.TriangleMesh.create_sphere(radius=0.1)
light_source.paint_uniform_color([1, 1, 1])
light_source.translate([2, 2, 2])

# Список названий моделей мебели
furniture_names = ["стул", "диван", "стол", "кровать", "дверь"]

# Преобразование каждой модели мебели в изображение
for furniture_name in furniture_names:
    for i in range(num_models):
        # Загрузка модели мебели из файла .obj
        file_name = f'{furniture_name}_{i}.obj'
        mesh = o3d.io.read_triangle_mesh(file_name)
        
        # Настройка параметров рендеринга
        vis = o3d.visualization.Visualizer()
        vis.create_window(width=image_width, height=image_height, visible=False)
        vis.add_geometry(mesh)
        vis.add_geometry(light_source)
        ctr = vis.get_view_control()
        #ctr.rotate(30.0, 0.0)
        vis.poll_events()
        vis.update_renderer()
        
        # Получение изображения и сохранение его в файле
        depth = vis.capture_depth_float_buffer(False)
        color = vis.capture_screen_float_buffer(False)
        color = np.asarray(color)
        color = (color * 255).astype(np.uint8)
        file_name = f'{furniture_name}_{i}.png'
        o3d.io.write_image(file_name, color)
```

```python
from gtts import gTTS
import os

# список названий моделей мебели
models = ["стул", "диван", "стол", "кровать", "дверь"]

# выбираем язык для озвучивания
language = 'ru'

for model in models:
    # создаем текст для озвучивания
    text = f"Модель мебели {model}"
    
    # создаем объект gTTS и передаем ему текст и язык
    tts = gTTS(text=text, lang=language)

    # сохраняем аудиофайл в текущей директории
    tts.save(f"{model}.mp3")

    # воспроизводим аудиофайл
    os.system(f"mpg123 {model}.mp3")
```

```python
import csv
import os

# список названий моделей мебели
models = ["стул", "диван", "стол", "кровать", "дверь"]

# название директории с изображениями
image_dir = "images"

# название директории с аудиофайлами
audio_dir = "audio"

# название CSV-файла
csv_filename = "dataset.csv"

# открываем файл для записи
with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # записываем заголовок файла
    writer.writerow(["image_filename", "audio_filename", "model_name"])
    
    # проходимся по каждой модели и добавляем соответствующую строку в файл
    for model in models:
        image_filename = os.path.join(image_dir, f"{model}.obj")
        audio_filename = os.path.join(audio_dir, f"{model}.mp3")
        writer.writerow([image_filename, audio_filename, model])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# загружаем датасет из CSV-файла
dataset = pd.read_csv('dataset.csv')

# определяем параметры разделения: 80% для обучающей выборки, 20% для проверочной
train_ratio = 0.8
test_ratio = 0.2

# разделяем данные на обучающую и проверочную выборки
train_data, test_data = train_test_split(dataset, test_size=test_ratio, train_size=train_ratio, shuffle=True)

# сохраняем данные в CSV-файлы
train_data.to_csv('train_dataset.csv', index=False)
test_data.to_csv('test_dataset.csv', index=False)
```
